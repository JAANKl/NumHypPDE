```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sc

y = np.linspace(-2, 2, 100)
t = 0.5/np.pi

def rhs(x, t, y):
    return np.sin(np.pi * x) * t + 0.5 * t + x - y

x = sc.newton(rhs, x0=y-t/2, args=(t, y))
plt.plot(y,x)
```

```python
t = 1.5/np.pi
x = sc.newton(rhs, x0=y-t/2, args=(t, y))
plt.plot(y,x)
```
